```python
import tkinter as tk
from tkinter import font
import requests
from PIL import Image, ImageTk


def format_response(weather):
    
    try:
        name = weather['name']
        desc = weather['weather'][0]['description']
        temp = weather['main']['temp']
        
        final_str = 'City: %s \n Conditions: %s\n Temperature: %s celsius' %(name, desc, temp)
    except:
        final_str = "No Records Found"
        
        
    return final_str
    

def know_weather(city):
    wkey = '09ca0094a71af04745fb433cfa3785de'
    link = 'https://api.openweathermap.org/data/2.5/weather'
    parameters = {'APPID': wkey, 'q': city, 'units': 'Metric'}
    response = requests.get(link, params = parameters)
    weather = response.json()
    
    label['text'] = format_response(weather)
    

root = tk.Tk()
root.title("Raja Weathering App")
canvas = tk.Canvas(root, height = 500, width = 600)
canvas.pack()

img = 'weather.jpg'
bgs = ImageTk.PhotoImage(file = img, master = root)
bgs_label = tk.Label(root, image = bgs)
bgs_label.place(relx = 0, rely = 0)

frame = tk.Frame(root, bg = '#003380', bd = 2.5)
frame.place(relx = 0.5, rely = 0.75, relwidth = 0.75, relheight = 0.1, anchor = 'n')

frame1 = tk.Frame(root, bg = '#003380', bd = 2.5)
frame1.place(relx = 0.5, rely = 0.1, relwidth = 0.75, relheight = 0.1, anchor = 'n')

entry = tk.Entry(frame1, font = ('Maiandra GD', 20))
entry.place(relx = 0.1, relheight = 1, relwidth = 0.8)

button = tk.Button(frame, text = "Show Weather", font = 40, command = lambda: know_weather(entry.get()))
button.place(relx = 0.06, relheight = 1, relwidth = 0.90)

#lf = Lower Frame
lf = tk.Frame(root, bg = '#009933', bd = 10)
lf.place(relx = 0.5, rely = 0.20, relwidth = 0.75, relheight = 0.55, anchor = 'n')

label = tk.Label(lf, font = ('Maiandra GD', 20))
label.place(relwidth = 1, relheight = 1)

root.mainloop()
```
